```python
import matplotlib.pyplot as plt
import src.pltparams as pltparams
import numpy as np


x = np.linspace(0, np.pi*2, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2*x)
y = y1 * y2

x_scatter = np.linspace(0, np.pi*2, 10)
y1_scatter = np.sin(x_scatter) + np.random.normal(0, 0.1, size=x_scatter.shape)
y2_scatter = np.cos(x_scatter) + np.random.normal(0, 0.1, size=x_scatter.shape)
y3_scatter = np.sin(2*x_scatter) + np.random.normal(0, 0.1, size=x_scatter.shape)
y_scatter = y1_scatter * y2_scatter

def plot_example(ax):
    ax.plot(x, y, label="data 1 line")
    ax.plot(x, y2, label = "data 2 line")
    ax.plot(x, y1, label = "data 3 line")
    ax.plot(x, y3, label = "data 4 line")

    ax.scatter(x_scatter, y_scatter, marker = next(pltparams.marker_cycle), label="data 1 scatter")
    ax.scatter(x_scatter, y2_scatter, marker = next(pltparams.marker_cycle), label="data 2 scatter")
    ax.scatter(x_scatter, y1_scatter, marker = next(pltparams.marker_cycle), label="data 3 scatter")
    ax.scatter(x_scatter, y3_scatter, marker = next(pltparams.marker_cycle), label="data 4 scatter")

    ax.set_xlabel(r"$X$-axis [Arbitrary units]")
    ax.set_ylabel(r"$Y$-axis [Arbitrary units]")

    ax.legend()
```

# Different lines, markers and colors

The defualt colorscheme has some hard to see colors for white backgrounds, especially if it defaults to the darkmode theme. The `'tab10'`colorshcme is good for white background, and is defualt here. Call `use_tab10_color` to revert back to it. 

For black and white, we call `use_classic_bw_style()` for different linestyles and markers instead of color. This is probably the most useful command.

There's also an alternate `use_okabe_ito_color()` if you prefer the okabe ito color scheme.

Cycle through the markers by simply calling `next(pltparams.marker_cycle)`. This is the easiest way, as setting default markers will also impose them on `plot()`in addition to `scatter()`. 

```python
# Plot single column (Default style )
fig, ax = plt.subplots(1,1,figsize = pltparams.figsize_single)
plot_example(ax)
fig.savefig("figures/example_single_default.png")
```

```python
pltparams.use_okabe_ito_color()
fig, ax = plt.subplots(1,1,figsize = pltparams.figsize_single)
plot_example(ax)
fig.savefig("figures/example_single_okabe_it.png")
```

```python
pltparams.use_classic_bw_lines()
fig, ax = plt.subplots(1,1,figsize = pltparams.figsize_single)
plot_example(ax)
fig.savefig("figures/example_single_bw.png")
```

## Different figure sizes

We have 2 version of all figure sizes; single column and double column. The widths can be accessed directly as `single_column_width` and `double_column_width` variables. 

The following default figure sizes are also available:
 - Default 1:1 ratio for double column and roughly 1:2 for double column
 - Golden ratio for double column and single column
 - Full page height for single and double column 

```python
# Plot double column
subplot_labels = ("first subplot", "second subplot", "third subplot")
for i,figsize in enumerate([pltparams.figsize_double, 
                          pltparams.figsize_double_golden, 
                          pltparams.figsize_double_fullpage]):
    fig, axs = plt.subplots(i+1,1,figsize = figsize)

    for j in range(i+1):
        if i > 0:
            ax = axs[j]
        else:
            ax = axs
        plot_example(ax)
        ax.set_title(pltparams.letter_labels_lc[j]+") : "+subplot_labels[j], loc='left')
    #fig.savefig(f"figures/example_double_{i+1}subplots.png")
```

```python
# Plot single column
sublabels = ('a)', 'b)', 'c)')
for i,figsize in enumerate([pltparams.figsize_single, 
                          pltparams.figsize_single_golden, 
                          pltparams.figsize_single_fullpage]):
    fig, axs = plt.subplots(i+1,1,figsize = figsize)

    for j in range(i+1):
        if i > 0:
            ax = axs[j]
        else:
            ax = axs
        plot_example(ax)
        ax.set_title(pltparams.letter_labels_lc[j]+") : "+subplot_labels[j], loc='left')    
```
